```python
from sklearn.preprocessing import StandardScaler
```

```python
# Data Cleaning
# -----No cleaning needed for this dataset(as it is previously cleaned data)------

# Data Integration
# -----Only one dataset is considered, so no integration is needed----

# Data Reduction 
# ------No reduction required, as there are considerable number of variables--------

# """"""Data Transformation""""""

# Normalize BMI using StandardScaler
scaler = StandardScaler()
bmi_scaled = scaler.fit_transform(data[['BMI']])
data['BMI_scaled'] = bmi_scaled
```

```python
# """"""Data Discretization""""""""""
#******** Discretize Age***************
def discretized_age(Age):
    if Age <= 18: 
        return 0
    elif Age <=39:
        return 1
    elif Age <= 64: 
        return 2
    else:
        return 3
data['age_discretized'] = data['Age'].apply(discretized_age)

#********* Discretize  Glucose  ************
def discretized_glucose(Glucose):  
    
    if Glucose < 100:
        return 0 #normal
    elif 100 <= Glucose <= 125:
        return 1 #pre-diabetic
    else:
        return 2 #diabetic

data['glucose_discretized'] = data['Glucose'].apply(discretized_glucose)

def discretized_bloodpressure(bp):
    if bp < 80:
        return 0  # low
    elif 80 <= bp <= 120:
        return 1  # normal
    else:
        return 2  # high

data['bloodpressure_discretized'] = data['BloodPressure'].apply(discretized_bloodpressure)

# ********* Discretize SkinThickness ************
def discretized_skinthickness(skin_thickness):
    if skin_thickness < 20:
        return 0  # thin
    elif 20 <= skin_thickness <= 30:
        return 1  # normal
    else:
        return 2  # thick

data['skinthickness_discretized'] = data['SkinThickness'].apply(discretized_skinthickness)

# ********* Discretize Insulin ************
def discretized_insulin(insulin):
    if insulin < 100:
        return 0  # low
    elif 100 <= insulin <= 150:
        return 1  # normal
    else:
        return 2  # high

data['insulin_discretized'] = data['Insulin'].apply(discretized_insulin)

# ********* Discretize BMI ************
def discretized_bmi(bmi):
    if bmi < 18.5:
        return 0  # underweight
    elif 18.5 <= bmi <= 24.9:
        return 1  # normal
    elif 25 <= bmi <= 29.9:
        return 2  # overweight
    else:
        return 3  # obese

data['bmi_discretized'] = data['BMI'].apply(discretized_bmi)

# ********* Discretize Diabetes Pedigree Function ************
def discretized_dpf(dpf):
    if dpf < 0.3:
        return 0  # low risk
    elif 0.3 <= dpf <= 0.6:
        return 1  # medium risk
    else:
        return 2  # high risk

data['dpf_discretized'] = data['DiabetesPedigreeFunction'].apply(discretized_dpf)
```
